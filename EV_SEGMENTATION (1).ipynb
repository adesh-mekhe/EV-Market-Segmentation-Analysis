```python
def EDA(df,figuresize):
    cat = []
    con = []
    from warnings import filterwarnings
    filterwarnings("ignore")
    import pandas as pd
    import seaborn as sb
    import matplotlib.pyplot as plt
    u = int((df.shape[1]+1)/2)
    plt.figure(figsize=figuresize)
    x = 1

    for i in df.columns:
        if(df[i].dtypes == "object"):
            cat.append(i)
            plt.subplot(u,2,x)
            x = x+1
            sb.countplot(df[i])
        else:
            con.append(i)
            plt.subplot(u,2,x)
            x = x+1
            sb.distplot(df[i])

    print(df.corr())
```

MODEL CREATION

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(df1)

pca = PCA(n_components=8)
pc = pca.fit_transform(pca_data)
names = ['var1','var2','var3','var4','var5','var6','var7','var8']
pf = pd.DataFrame(data = pc,columns = names)
```

```python
pf
```

```python
#Proportion of Variance (from var1 to var8)
pca.explained_variance_ratio_
```

```python
import numpy as np
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# correlation coefficient between original variables and the component

loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["var"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df1.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
import matplotlib.pyplot as plt
```

```python
#Correlation matrix plot for loadings
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
#K-means clustering

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(pf)
df['cluster_num'] = kmeans.labels_ #adding to df
print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares.
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_) #Location of the centroids on each cluster.
```

```python
#Visulazing clusters
sns.scatterplot(data=pf, x="var1", y="var2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

selecting the segments

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['Used it for'] = LabelEncoder().fit_transform(df['Used it for'])
visit = df.groupby('cluster_num')['Used it for'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
df['Used it for'] = LabelEncoder().fit_transform(df['Used it for'])
used = df.groupby('cluster_num')['Used it for'].mean()
used = used.to_frame().reset_index()
used
```

```python
df['Model Name'] = LabelEncoder().fit_transform(df['Model Name'])
diff_models= df.groupby('cluster_num')['Model Name'].mean()
diff_models = diff_models.to_frame().reset_index()
diff_models
```

```python
segments =used.merge(diff_models, on='cluster_num', how='left')
segments
```

```python
#Target segments

plt.figure(figsize = (9,4))
sns.scatterplot(x = "Used it for", y = "Model Name",data=segments,s=400, color="r")
plt.title("Simple Behavioural_segmentation for EV data",
          fontsize = 15)
plt.xlabel("Used it for", fontsize = 12)
plt.ylabel("Model Name", fontsize = 12)
plt.show()
```
